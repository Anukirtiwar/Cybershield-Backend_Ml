```python
from paddleocr import PaddleOCR

# âœ… Use multilingual model (supports Hindi + English + 80+ languages)
ocr_multi = PaddleOCR(use_angle_cls=True, lang='en')  # for English + Hindi use lang=''

def extract_multilingual_text(image_path):
    results = ocr_multi.ocr(image_path, cls=True)

    extracted_text = []
    for line in results[0]:
        extracted_text.append(line[1][0])  # line[1][0] = recognized text

    return " ".join(extracted_text)

# Example usage
if _name_ == "_main_":
    img_path = "dr1.png"  # ðŸ”„ change to your image path
    output = extract_multilingual_text(img_path)

    print("âœ… Extracted Text (English + Hindi):")
    print(output)
```
